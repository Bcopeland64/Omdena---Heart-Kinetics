<h1>Classification of Metadata using Pycaret</h1>


```python
from pycaret.classification import *
```

```python
exp_1 = setup(data = records, target = 'hf_type', session_id=123)
```

```python
#Compare Top 5 Models
top_5 = compare_models(n_select = 5)
blender = blend_models(estimator_list = top_5, fold = 5, method = 'soft')
```

```python
#Stack Top 5 Models for better performance
stacker_1 = stack_models(estimator_list = top_5, meta_model = blender)
```

```python
#Plot Stacked Model
evaluate_model(stacker_1)
```

```python
#Predict on test / hold-out Sample
predict_model(stacker_1)
```

```python
#Finalize Stacked Model for Deployment
finalize_model(stacker_1)
```

```python
#Save Model for Later Use
save_model(stacker_1, 'stacked-model-1')
```
